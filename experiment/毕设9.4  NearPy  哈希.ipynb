## 官方示例

```python
import numpy

from nearpy import Engine
from nearpy.hashes import RandomBinaryProjections
from nearpy.filters import NearestFilter
from nearpy.distances import CosineDistance
import time
```

```python
# Dimension of our vector space
dimension = 512

# Create a random binary hash with 10 bits
rbp1 = RandomBinaryProjections('rbp', 6)
rbp2 = RandomBinaryProjections('rbp', 6)
rbp3 = RandomBinaryProjections('rbp', 6)

# Create engine with pipeline configuration
engine = Engine(dimension, lshashes=[rbp1,rbp2,rbp3], distance = CosineDistance(), vector_filters=[NearestFilter(40)])

# Index 1000000 random vectors (set their data to a unique string)
for index in range(400):
    v = numpy.random.randn(dimension)
    engine.store_vector(v, index)

# Create random query vector
query = numpy.random.randn(dimension)

time_start = time.time()
# Get nearest neighbours
N = engine.neighbours(query)
time_end = time.time()
print('Time cost = %fs' % (time_end - time_start))
```
